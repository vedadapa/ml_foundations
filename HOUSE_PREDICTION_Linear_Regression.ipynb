# HOUSE PREDICTION

*   weights and bias

*   HYPER TUNING - Learning_rate and epochs

*   price prediction using the sqft

*   Mean Squared Error

*   





```python
import numpy as np

# Simple dataset - house sizes and prices
X = np.array([1, 2, 3, 4, 5])  # size in 1000 sqft
y = np.array([150, 200, 250, 300, 350])  # price in lakhs
weight = 0
bias = 0

# hyperparameter tuning
learning_rate = 0.01
epochs = 5000

n = len(X)
# each life cycle calculation 5000 epochs
for epoch in range(epochs):

# claculating predicted values per sqft
  y_pred = (weight * X) + bias

# identifying the mean squared error to predicted values
  MSE = np.mean((y - y_pred) ** 2)

# direction for weight and bias
  dw = (-2/n) * np.sum(X*(y-y_pred))
  db = (-2/n) * np.sum(y - y_pred)

# gradient tuning
  weight = weight - learning_rate * dw
  bias = bias - learning_rate * db

# the slope of the gradiant
  if epoch % 100 == 0:
    print(f" epoch: {epoch} and loss: {MSE}")

print('final weights:',weight)
print("final bias:",bias)





# Your job - implement these from scratch:
# 1. Initialize
# 2. Write predict() function: y = (weight * X) + bias
# 3. Write mse() function: mean squared error
# 4. Print initial predictions and initial MSE
```
